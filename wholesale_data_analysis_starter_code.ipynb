## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
df4=df[["unit_price","qty"]]
#df4
#df4["qty"]=pd.to_numeric(df4["qty"], errors='coerce').fillna(0).astype(int)
#df4['qty'] = pd.to_numeric(df4['qty'], errors='coerce').fillna(0).astype(int)
#df4
df.loc[:, "qty"] = pd.to_numeric(df["qty"], errors="coerce").fillna(0).astype(int)
df.loc[:, "unit_price"] = pd.to_numeric(df["unit_price"], errors="coerce")
df.loc[:,"line_subtotal"]=df.loc[:,"unit_price"]*df.loc[:,"qty"]
df[["unit_price","qty", "line_subtotal"]].head(2)
df
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.
df.loc[:, "unit_weight"] = pd.to_numeric(df["unit_weight"], errors="coerce") #.fillna(0).astype(float)
df.loc[:, "unit_price"] = pd.to_numeric(df["unit_price"], errors="coerce")
df.loc[:,"total_weight"]=df.loc[:,"unit_weight"]*df.loc[:,"qty"]
def cal_shipping_price(weight):
    if weight > 50:
        return weight * 7
    return weight * 10
df.loc[:,"shipping_price"]=df["total_weight"].apply(cal_shipping_price)
df[["unit_weight","qty", "total_weight","shipping_price"]].head(2)
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%
#df.loc[:,"line_price"]=round((df.loc[:,"line_subtotal"]+df.loc[:,"shipping_price"])*1.0925,2) gives an error
df.loc[:,"line_price"]=(df.loc[:,"line_subtotal"]+df.loc[:,"shipping_price"])*1.0925
df[["line_subtotal","shipping_price","line_price"]].head(2)
#df["line_price"]= df["line_price"].round(2) gives an error
df["line_price"]= df["line_price"].apply(lambda x: round(x, 2))
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).
df.loc[:,"line_cost"]=(df.loc[:,"unit_cost"]*df.loc[:,"qty"])+df.loc[:,"shipping_price"]
df.head(3)
```

```python
# Create a column for the profit of each line using line cost and line price

df.loc[:,"line_profit"]=(df.loc[:,"line_price"]-df.loc[:,"line_cost"])
df.head(3)
```

Use the new columns with confirmed values to find the following information.

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit. 
df_grouped2=df.groupby("client_id").agg({"line_price":"sum", "qty":"sum", "shipping_price":"sum", "line_price":"sum", "line_cost":"sum", "line_profit":"sum"})
client_ids = [33615, 66037, 46820, 38378, 24741]

df_summary = df_grouped2[df_grouped2.index.isin(client_ids)]
df_summary
```

```python
# Format the data and rename the columns to names suitable for presentation.
df_summary2 = df_summary.reset_index(drop=False)
df_summary3=df_summary2.rename(columns={
    'client_id': 'Client ID',
    'qty': 'Units',
    'shipping_price': 'Shipping',
    'line_price': 'Total Revenue',
    'line_cost': 'Total Cost',
    'line_profit': 'Total Profit'
})

# Define the money columns. 

money_columns = ['Shipping', 'Total Revenue', 'Total Cost', 'Total Profit']

# Define a function that converts a dollar amount to millions.
def convert_to_millions(dolar):
    return dolar / 1000000

# Apply the currency_format_millions function to only the money columns. 
df_summary3[money_columns] = df_summary3[money_columns].map(convert_to_millions)

# Rename the columns to reflect the change in the money format. 

df_summary4=df_summary3.rename(columns={
    
    'Shipping': 'Shipping (millions)',
    'Total Revenue': 'Total Revenue (millions)',
    'Total Cost': 'Total Cost (millions)',
    'Total Profit': 'Total Profit (millions)'
})

df_summary4.head(5)
```

```python
# Sort the updated data by "Total Profit (millions)" form highest to lowest and assign the sort to a new DatFrame.
df_summary5=df_summary4.sort_values('Total Profit (millions)', ascending = False)

df_summary5.head(5)
```
